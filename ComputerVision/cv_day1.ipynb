```python
color_list = [method for method in dir(cv2) if method.startswith('COLOR_') is True]

img = cv2.imread('/Users/pawankumarkc/Downloads/Datasets_to_delete/CV/images/opencv.jpg')

for color in range(len(color_list)):
    col_name = color_list[color]
    col_code = 'cv2.'+col_name
    resized = cv2.resize(img, (400,400))
    try:
        rgb_img = cv2.cvtColor(resized, eval(col_code))
        cv2.imshow(col_code+"-"+str(color), rgb_img)
        cv2.waitKey()
    except Exception as e:
        print('Error on '+str(e))
        pass
cv2.destroyAllWindows()
```
