## Percptron model - Learning purposes only

```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.datasets import load_digits


##Yann lecun
```

```python
digits = load_digits()
digits
```

```python
digits['data']
```

```python
digits['target']
```

```python
x = digits.data
y = digits.target

#This is a multi class classification problem
```

```python
print(x.shape)
print(y.shape)
```

```python
#Split the data into train and test

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=1)
```

```python
#x = x/255.0
```

```python
# Building Perceptron modesl
#Scaling is required.
from sklearn.linear_model import Perceptron

perceptron_model = Perceptron()
perceptron_model.fit(x_train, y_train)
```

```python
print("What is the training accuracy of perceptron model")
print(perceptron_model.score(x_train, y_train))

print("What is the test accuracy of perceptron model")
print(perceptron_model.score(x_test, y_test))

perceptron_train = perceptron_model.predict(x_train)
perceptron_test = perceptron_model.predict(x_test)

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
print("Training accuracy", accuracy_score(y_train, perceptron_train))
print("Test accuracy",accuracy_score(y_test, perceptron_test))

print("Classification report Train")
print(classification_report(y_train, perceptron_train))
print("Classification report Test")
print(classification_report(y_test, perceptron_test))
```

## Regression problem

```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense, Dropout, BatchNormalization
```

```python
data = pd.read_csv('/Users/pawankumarkc/Documents/vscode_workspace/MLAlgo/datasets/boston.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
x = data.iloc[:,:-1]
y = data.iloc[:,-1]

x.head()
```

```python
y.head()
```

```python
## Feature scaling - mandatory step
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x = sc.fit_transform(x)
pd.DataFrame(x).head()
```

# Building multilayer perceptron - DNN


```python
# Building multilayer perceptron - DNN
dnn = Sequential()
#Add hidden layer 1
dnn.add(Dense( 8, activation='relu', input_dim=13))
#Batch normalisation
dnn.add(BatchNormalization())
#Dropout
dnn.add(Dropout(0.25))
# Add one more hidden layer
dnn.add(Dense( 8, activation='relu'))
dnn.add(BatchNormalization())
dnn.add(Dropout(0.25))
#output
dnn.add(Dense( 1, activation='linear'))

dnn.summary()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=101)
```

```python
# Compile the model using optimization and loss function

dnn.compile(optimizer='sgd', loss='mean_squared_error')

#Fit the model
history  = dnn.fit(x_train, y_train, epochs=500)
```

```python
# Prediction test dataset

y_pred = dnn.predict(x_test)
y_pred
```

```python
print(y_pred[0:10])
print('************************')
print(y_test[0:10])
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test, y_pred))
```
