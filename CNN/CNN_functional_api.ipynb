## Case 1: Input =1 and Output = 2

```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Model
from keras.layers import *
```

```python
# Input = 1
x = Input(shape=(3,))
```

```python
x
```

```python
hidden1 = Dense(64, activation='relu')(x) #Here x is input shape.  
hidden2 = Dense(64, activation='relu')(hidden1) #Output of hidden1 is connected to hidden 2
hidden3 = Dense(64, activation='relu')(hidden2)

output1 = Dense(1, activation='linear')(hidden3)     #Connecting last hidden layer to output layer. Case linear regression
output2 = Dense(1, activation='sigmoid')(hidden3)   #Case binary classification
output3 = Dense(10, activation='softmax')(hidden3)  #Case multiclass classification
```

```python
model = Model(inputs = x, outputs = [output1, output2, output3])    #Building multiple approaches in a single model
model.summary()
```

## Case 2: Input = 3 and Output = 1

```python
input1 = Input(shape=(32,))
input2 = Input(shape=(64,))
input3 = Input(shape=(128,))
```

```python
#The first branch operates on the first dataset
#Connected with 3 hidden layer
x = Dense(16, activation='ReLU')(input1)
x1 = Dense(24, activation='ReLU')(x)
x2 = Dense(64, activation='ReLU')(x1)

#The second branch operates on the second dataset
#Connected with 2 hidden layer
y = Dense(16, activation='ReLU')(input2)
y1 = Dense(128, activation='ReLU')(y)

#The third branch operates on the third dataset
#Conneced with 4 hidden layer
z = Dense(128, activation='ReLU')(input3)
z1 = Dense(24, activation='ReLU')(z)
z2 = Dense(24, activation='ReLU')(z1)
z3 = Dense(24, activation='ReLU')(z2)
```

```python
#Combining all outputs of the 3 branches

combined = concatenate([x2, y1, z3])
```

```python
#Applying a fully connected layer and then a regression prediction on the combined output

m = Dense(32, activation='ReLU')(combined)
m1 = Dense(1, activation='linear')(m)

#Connect with functional API
model = Model(inputs=[input1, input2, input3], outputs=m1)
model.summary()
```

```python
plot_model(model, show_shapes=True)
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from keras.preprocessing.image import ImageDataGenerator
```

```python
# Predict age(Regression), gender(Binary classification) and ethnicity(Multiclass Classification)

age = []
gender = []
ethnicity = []
image_path = []

for file in os.listdir(folder_path):
    age.append(int(file.split('_')[0]))
    gender.append(int(file.split('_')[1]))
    #ethnicity.append(int(file.split('_')[2]))
    image_path.append(file)
```

```python
print(age)
print(len(age))

print(gender)
print(len(gender))
```

```python
df = pd.DataFrame({'age': age, 'gender': gender, 'img_path': image_path})
df.head()
```

```python
df.shape
```

```python
#Split the data into training and validation

train_df = df.sample(frac=1, random_state=0).iloc[:20000]
test_df = df.sample(frac=1, random_state=0).iloc[20001:]

print(train_df.shape, test_df.shape)
```

```python
#Data augumentation
train_datagen = ImageDataGenerator(rescale= 1./255, rotation_range= 25, 
                                   width_shift_range=0.25, height_shift_range=0.2, zoom_range=0.2, 
                                   horizontal_flip=True)

test_datagen = ImageDataGenerator(rescale= 1./255, rotation_range= 25, 
                                   width_shift_range=0.25, height_shift_range=0.2, zoom_range=0.2, 
                                   horizontal_flip=True) #Only rescale is sufficient for test data
```

```python
# Apply data augumentation with original dataset
train_generator = train_datagen.flow_from_dataframe(train_df, directory=folder_path, x_col='img_path', 
                                                    y_col=['age', 'gender'], target_size=(256, 256), class_mode='multi_output')
#Specifying class_mode='multi_output' since we have multiple things to predict

test_generator = train_datagen.flow_from_dataframe(test_df, directory=folder_path, x_col='img_path', 
                                                    y_col=['age', 'gender'], target_size=(256, 256), class_mode='multi_output')

```

### Resnet transfer learning

```python
#Resnet model

from keras.layers import *
from keras.models import Model
from keras.applications.resnet50 import ResNet50
```

```python
resnet = ResNet50(include_top=False, input_shape=(256, 256, 3))
```

```python
resnet = ResNet50(include_top=False, input_shape=(256, 256, 3))
resnet.trainable = False

output = resnet.layers[-1].output
flatten = Flatten()(output)

#Creating 2 dense layers since we have both regression and classification problems
dense1 = Dense(128, activation='relu')(flatten)
dense2 = Dense(128, activation='relu')(flatten)

dense3 = Dense(64, activation='ReLU')(dense1)
dense4 = Dense(64, activation='ReLU')(dense2)


output1 = Dense(1, activation='linear', name='age')(dense3) #specifing name='age' since we are using output1 for predicting age
output2 = Dense(1, activation='sigmoid', name='gender')(dense4)

model = Model(inputs=resnet.input, outputs=[output1, output2])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss={'age': 'mae', 'gender':'binary_crossentropy'},
               metrics={'age': 'mae', 'gender':'accuracy'}, loss_weights={'age': 10, 'gender':90})
```

```python
history = model.fit(train_generator, batch_size=32, epochs=5, validation_data=test_generator)
```
