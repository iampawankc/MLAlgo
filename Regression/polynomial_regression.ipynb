```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.metrics import r2_score
```

```python
x = 6 * np.random.rand(200,1)-3
y = 0.8 * x**2 +0.9 * x + 2 +np.random.randn(200,1)
```

```python
x.shape
```

```python
y.shape
```

```python
plt.plot(x,y, 'b')
plt.xlabel('X datapoint')
plt.xlabel('Y datapoint')
```

```python
pd.DataFrame(x).describe
```

```python
# Split the data into train and test 

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=2, train_size=0.2)
```

```python
#Linear regression model 

lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
r2_score(y_test, y_pred)
# R2 score is very bad
```

```python
plt.plot(x_train, lr.predict(x_train), color='r')
plt.plot(x, y, color='b')
plt.xlabel('X datapoint')
plt.xlabel('Y datapoint')
```

```python
# If data in non liner, use polinomial regression model 

poly = PolynomialFeatures()
x_train_trans = poly.fit_transform(x_train)
x_test_trans = poly.fit_transform(x_test)
```

```python
#linear regression
lr = LinearRegression()
lr.fit(x_train_trans, y_train)
```

```python
y_pred = lr.predict(x_test_trans)
```

```python
r2_score(y_test, y_pred)
#Accuracy increase drastically
```

```python
y_pred_train = lr.predict(x_train_trans)
r2_score(y_train, y_test)
```

```python
plt.plot(x_train_trans, lr.predict(x_train_trans), color='r')
plt.plot(x, y, color='b')
plt.xlabel('X datapoint')
plt.xlabel('Y datapoint')
```
