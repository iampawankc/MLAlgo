```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60]
y=[5,10,15,20,25,30]

plt.plot(x,y,color='red', ls='-', lw=3, marker='<', markerfacecolor='yellow', markersize=6,
         markeredgewidth=2, markeredgecolor='g', label='X vs Y')
plt.plot(y,x, color='blue')
plt.plot(x*2, y*2, color='black')
plt.plot([i^2 for i in x], [i^2 for i in y], color='orange')
plt.title('Plot-1', fontdict={'font':'monospace', 'color':'red'})
plt.xlabel('Array X')
plt.ylabel('Array Y')
plt.legend(loc=0, labelcolor='b', fontsize=10)

plt.show()
```
